OpenCV Provide Many preytrained Model To detect Objetcs Faces ....

```python
import cv2
import matplotlib.pyplot as plt 
```

```python
faceCascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
eyeCascadde = cv2.CascadeClassifier("haarcascade_eye.xml")
img= cv2.imread("../images/mee++.jpg",1) # Gray 
#img =cv2.resize(img,(400,400))
print(f"Height : {img.shape[0]} , Width : {img.shape[1]}")

## Fun ction To Detect Face 
def detect_face(image) :
    img = image.copy()
    face_rect = faceCascade.detectMultiScale(img,scaleFactor=5,minNeighbors=5)
    for (x,y,w,h) in face_rect :
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,255),3)

    return img

def detect_eye(image):
    img = image.copy()
    eye_rect = eyeCascadde.detectMultiScale(img,scaleFactor=2,minNeighbors=5)

    for (x,y,w,h) in eye_rect :
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)
    return img
```

You Better Scale Up your Image :

```python
w= 500
ratio = img.shape[0]/img.shape[1] # Height / Width
h= int(ratio * w)

img_resized = cv2.resize(img,(w,h))
eye_face_resized = detect_eye(detect_face(img_resized))
eye_face = detect_eye(detect_face(img))

# Show the images using Matplotlib
eye_face_RGB = cv2.cvtColor(eye_face,cv2.COLOR_BGR2RGB)
eye_face_resized_RGB = cv2.cvtColor(eye_face_resized,cv2.COLOR_BGR2RGB)

fig ,axes = plt.subplots(1,2,figsize=(10,10))
axes[0].imshow(eye_face_resized_RGB) ,axes[0].title.set_text("Face Detection"), axes[0].axis("off")
axes[1].imshow(eye_face_RGB) ,axes[1].title.set_text("Face Not Resized") , axes[1].axis("off")

plt.show()

# Uncomment To See The Result 
#cv2.imshow("Face Detection",eye_face_resized)
#cv2.imshow("Face Not Resized",eye_face)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```
